```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
from datetime import datetime
```

```python
msft = yf.download("msft", start="2024-07-15", end="2024-07-19", interval="1m")
msft['id'] = np.arange(len(msft))

xticks_name = [
    "2024-07-15 09:30:00-04:00",
    "2024-07-16 09:30:00-04:00",
    "2024-07-17 09:30:00-04:00",
    "2024-07-18 09:30:00-04:00",
    "2024-07-18 15:59:00-04:00"
]
xticks_id = msft.loc[xticks_name]['id'].to_list()
xticks_name = [f"{v.split()[0]}\n{v.split()[1].split(":")[0]}:{v.split()[1].split(":")[1]}" for v in xticks_name]

plt.figure(figsize=(8, 8))
plt.plot(np.arange(len(msft)), msft.Close, color="red", alpha=0.8, linewidth=1)
plt.fill_between(np.arange(len(msft)), np.full(len(msft), msft.Close.min()), msft.Close, color="red", alpha=0.1)
plt.xticks(xticks_id, xticks_name, fontsize=12)
plt.yticks(np.arange(435, 456, 5), [f"${v}" for v in np.arange(435, 456, 5)], fontsize=12)
plt.title("Microsoft stock prices", fontsize=21)
plt.show()
```

```python
msft = yf.download("msft", start="2024-07-15", end="2024-07-19", interval="1m")
msft['id'] = np.arange(len(msft))

xticks_name = [
    "2024-07-15 09:30:00-04:00",
    "2024-07-16 09:30:00-04:00",
    "2024-07-17 09:30:00-04:00",
    "2024-07-18 09:30:00-04:00",
    "2024-07-18 15:59:00-04:00"
]
xticks_id = msft.loc[xticks_name]['id'].to_list()
xticks_name = [f"{v.split()[0]}\n{v.split()[1].split(":")[0]}:{v.split()[1].split(":")[1]}" for v in xticks_name]

plt.figure(figsize=(8, 8))
plt.plot(np.arange(len(msft)), msft.Close, color="red", alpha=0.8, linewidth=1)
plt.fill_between(np.arange(len(msft)), np.zeros(len(msft)), msft.Close, color="red", alpha=0.1)
plt.xticks(xticks_id, xticks_name, fontsize=12)
plt.yticks(np.arange(50, 451, 50), [f"${v}" for v in np.arange(50, 451, 50)], fontsize=12)
plt.ylim(0, msft.Close.max()*1.1)
plt.title("Microsoft stock prices", fontsize=21)
plt.show()
```
