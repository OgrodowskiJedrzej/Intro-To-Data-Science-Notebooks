```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X = np.arange(50)
y = np.random.normal(size=50)
y = np.array([np.mean(y[max(i-5, 0): i+1]) for i in range(50)])
y = np.linspace(10, 100, 50) + y * 20
y
```

```python
plt.figure(figsize=(7, 6))
plt.plot(X, y, color="red")
plt.fill_between(X, y, alpha=0.3, color="red")
plt.xticks([])
plt.savefig("./imgs/vis_after.png", dpi=200)
plt.show()
```
