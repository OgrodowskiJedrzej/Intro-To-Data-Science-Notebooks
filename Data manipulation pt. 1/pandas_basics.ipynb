# Aggregation

```python
data = pd.read_csv("./data/titanic.csv")
```

#### `.mean`, `.count`, `.sum`, etc.

```python
data[["Sex", "Age"]].groupby(["Sex"]).mean()
```

```python
data[["Sex", "PassengerId"]].groupby("Sex").count()
```

```python
data[["Sex", "Fare"]].groupby("Sex").sum()
```

#### `.agg`

```python
data[["Sex", "Fare"]].groupby("Sex").agg(["sum", "mean", "count"])
```
